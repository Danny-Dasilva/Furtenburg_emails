```
#@title Send Multiple Emails
import csv
from collections import defaultdict
import os
import smtplib
from email.message import EmailMessage 
from email.utils import make_msgid

csv_file = list(uploaded.keys())[0] 
columns = {}
with open(csv_file) as f:
    reader = csv.DictReader(f) # read rows into a dictionary format
    for row in reader: # read a row as {column1: value1, column2: value2,...}
        for (k,v) in row.items(): # go over each column name and value 
            columns.setdefault(k,[]).append(v) # append the value into the appropriate list
                                 # based on column name 


import mimetypes
sender = 'ddasilva@furtenburg.com' #@param {type:"string"}
password = 'awtf#1662' #@param {type:"string"}


def send_message(receiver):
  msg = EmailMessage()

  msg['From'] = sender
  msg['To'] = receiver
  Email_Subject = 'Furtenburg Capital Lending Group- PPP Programs, Loans & Lines of Credit' #@param {type:"string"}
  msg['Subject'] = Email_Subject
  msg.add_alternative(footer, subtype='html')     
  with smtplib.SMTP_SSL('smtpout.secureserver.net', 465) as server:
    server.login(sender, password)
    server.send_message(msg)

for val in columns['Email']:
  try:
    send_message(val)
    print(f'message sent to: {val}')
  except:
    print('error, something went wrong')
```
